```python
import numpy as np
import glob
import cv2
import os
import natsort
```

```python
img_array = []
filelist = sorted(glob.glob('/home/user/GAPHeat/GC/figure/exa/*'))
filelist=natsort.natsorted(filelist)

for filename in filelist:
    img = cv2.imread(filename)
    h, w, c = img.shape
    size = (w, h)
    img_array.append(img)

out = cv2.VideoWriter('exa.avi',cv2.VideoWriter_fourcc(*'DIVX'), 20, size)

for i in range(len(img_array)):
    out.write(img_array[i])
    
out.release()
    
```
